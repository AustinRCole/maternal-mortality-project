```python
import pandas as pd
```

```python
df = pd.read_csv("../Resources/non_race_data.csv")
```

```python
df.head()
```

```python
# Create the bins for grouping MMR data
  
bins = [0, 4.99, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 44.99, 49.99, 54.99, 59.99, 64.99, 69.99, 74.99, 79.99, 84.99, 89.99, 94.99, 100]

# Create the names for the bins
mmr_groups = ["<5", "5-9.99", "10-14.99", "15-19.99", "20-24.99", "25-29.99", "30-34.99", "35-39.99", 
               "40-44.99", "45-49.99", "50-54.99", "55-59.99", "60-64.99", "65-69.99", "70-74.99", "75-79.99", "80-84.99", "85-89.99", "90-94.99", "95+"]
```

```python
df["mmr_group"] = pd.cut(df["maternal_mortality_ratio"], bins, labels=mmr_groups, include_lowest=True)
df
```

```python
X = df.drop(["year", "state", "id", "state_code", "latitude", "longitude", "deaths", "births", "maternal_mortality_ratio", "mmr_group"], axis=1)
y = df["mmr_group"]
print(X.shape, y.shape)
```

```python
# Split data into train and test groups
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Scale the data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Logistic Regression

```python
# Make logistic regression model
from sklearn.linear_model import LogisticRegression
LoR_model = LogisticRegression()
LoR_model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {LoR_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {LoR_model.score(X_test_scaled, y_test)}")
```

```python
predictions = LoR_model.predict(X_test_scaled)
print(f"First 10 Predictions:   {predictions[:20]}")
print(f"First 10 Actual labels: {y_test[:20].tolist()}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=900)
rf_model = rf_model.fit(X_train_scaled, y_train)
rf_model.score(X_test_scaled, y_test)
```

```python
feature_importances = pd.DataFrame(rf_model.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
```
