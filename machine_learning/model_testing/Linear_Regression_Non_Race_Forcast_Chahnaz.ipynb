### Performing Time-Series Forecast of Average Annual predictions of all other variables in USA (2020-2030)

#### Simple Time-Series Forecast of Average Annual Maternal Mortality in USA 

```python
# Importing ML linear regression dependency
from sklearn.linear_model import LinearRegression

# Selecting and assigning the X and y values
X1 = AVG_MMR_USA.reset_index().year.values.reshape(-1,1)
y1 = AVG_MMR_USA.reset_index()['maternal_mortality_ratio'].values

# Steps 1 & 2: Creating and Fitting our model to the data
reg_model = LinearRegression().fit(X1, y1)

# Obtaining the coefficient of determination (R2)
r_square = reg_model.score(X1, y1)

# Setting predictions dictionary to hold variables
predictions=[]

# Creating a for-loop to return predictions from model
for i in range(2020,2031):
    y_values=reg_model.predict(np.array([i]).reshape(1, -1))
    predictions.append(y_values[0])

# Creating a year range to hold years values with the future predictions
years = [i for i in range(2020,2031)] 

# Creating a new DataFrame to hold mmmr values with 10 years predictions
mmr_predict_df = pd.DataFrame(list(zip(years, predictions)), 
               columns =['year', 'maternal_mortality_ratio']) 


# DataFrame with the future predictions
mmr_predict_df
```

```python
# Selecting y value
y = 'diabetes_val'
# Grouping by year
AVG_MMR_USA_diabetes = mmr_without_race.groupby('year').mean()[y]

# Selecting and assigning the X and y values
X2 = AVG_MMR_USA_diabetes.reset_index().year.values.reshape(-1,1)
y2 = AVG_MMR_USA_diabetes.reset_index()['diabetes_val'].values

# Creating and Fitting our model to the data
model2 = LinearRegression().fit(X2, y2)
r_sq_2 = model2.score(X2, y2)
# print(r_sq_2)

# Setting predictions dictionary to hold variables
diabetes=[]

# Creating a for-loop to return predictions from model
for i in range(2020,2031):
    y_values = model2.predict(np.array([i]).reshape(1, -1))
    diabetes.append(y_values[0])
    
# Creating a year range to hold years values with the future predictions
years = [i for i in range(2020,2031)] 

# Creating a new DataFrame to hold values with 10 years predictions
predictions2 = pd.DataFrame(list(zip(years, diabetes)), 
               columns =['year', 'diabetes_val']) 

# DataFrame with the future predictions
predictions2

# Merging with mmr_predictions 
combined1 = pd.merge(mmr_predict_df, predictions2, how='inner', left_on=['year'], right_on=['year'])
combined1
```
