### Importing Dependencies

```python
# Importing dependencies:
import numpy as np
import pandas as pd
```

### Read the CSV and Perform Basic Data Cleaning

```python
# Reading the csv file into a pandas DataFrame
mmr_without_race = pd.read_csv('../Resources/non_race_data.csv')
mmr_without_race
```

### Grouping DataFrame by `year` and Calculating the Average

```python
# Setting y value for grouping
y = 'maternal_mortality_ratio'

# Grouping by year
AVG_MMR_USA = mmr_without_race.groupby('year').mean()[y]
AVG_MMR_USA.reset_index()
```

```python
# Setting y value for grouping
y = 'prem_death_val'

# Grouping by year
AVG_MMR_USA_1 = mmr_without_race.groupby('year').mean()[y]
AVG_MMR_USA_1.reset_index()

# Merging with mmr_predictions 
df_1 = pd.merge(AVG_MMR_USA, AVG_MMR_USA_1, how='inner', left_on=['year'], right_on=['year'])
df_1.reset_index()
```

```python
# Setting y value for grouping
y = 'diabetes_val'

# Grouping by year
AVG_MMR_USA_2 = mmr_without_race.groupby('year').mean()[y]
AVG_MMR_USA_2.reset_index()

# Merging with mmr_predictions 
df_2 = pd.merge(df_1, AVG_MMR_USA_2, how='inner', left_on=['year'], right_on=['year'])
df_2.reset_index()
```
