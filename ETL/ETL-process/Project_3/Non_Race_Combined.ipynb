### Create new dataframes with select columns

```python
# Filtering dataset by selecting columns needed
CDC_MMR_subset = CDC_MMR[["state", "year", "deaths", "births", "maternal_mortality_ratio", "population"]]
CDC_MMR_subset.head()
```

```python
# Filtering dataset by selecting columns needed
US_Ins_subset = US_Ins[["location", "year", "employer", "non_group", "medicaid", "medicare", "military", "uninsured"]]
US_Ins_subset.head()
```

```python
# Renaming the subset dataframe column from location to state to join on
US_Ins_df = US_Ins_subset.rename(columns={'location': 'state'})
US_Ins_df.head()
```

### Combine Dataframes on `state` and `year` Columns

```python
# Combining the dataframes on `state` and `year` columns
combined_df1 = pd.merge(states_df, CDC_MMR_subset, how='left', left_on=['state', 'year'], right_on=['state', 'year']).dropna()
combined_df1
```

```python
# Combining the `combined_df1` to `US_Ins_df` on state` and `year` columns
combined_df2 = pd.merge(combined_df1, US_Ins_df, how='inner', left_on=['state', 'year'], right_on=['state', 'year']).dropna()
combined_df2
```

```python
# Combining Final DataFrame
final_combined_df = pd.merge(combined_df2, State_Values, how='inner', left_on=['state', 'year'], right_on=['state', 'year']).dropna()
final_combined_df
```

```python
# Getting list of all column names
for col in final_combined_df.columns:
    print(col)
```

```python
# Saving merged dataframe into a csv file, without the Pandas Index, but with the header
final_combined_df.to_csv("../../../machine_learning/Resources/non_race_data2.csv", index=False, header=True, float_format='%.2f')
```
