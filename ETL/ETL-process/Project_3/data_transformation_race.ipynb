```python
# Dependencies
import pandas as pd
```

## State Level (excluded race)

```python
filepath_mmr_race = "../../output_file/INS_MMR_by_Race_ML.csv"
filepath_state_info = "../../output_file/state_non_race_info.csv"
race_df = pd.read_csv(filepath_mmr_race)
state_df = pd.read_csv(filepath_state_info)
```

```python
race_df.head()
```

```python
race_df.columns
```

```python
print(len(state_df))
state_df.head()
```

```python
cleaned_race_df = race_df[['state', 'state_code', 'year', 'race','hispanic_origin',
        'births_by_race', 'deaths_by_race', 'mmr_by_race', 'population_by_race']].copy()

print(len(cleaned_race_df))
cleaned_race_df.head()
```

```python
final_race_df = cleaned_race_df.merge(state_df, how='left', on=['state', 'year'])
```

```python
print(len(final_race_df))
final_race_df.head()
```

```python
final_race_df.rename(columns={
    'state_code_x': 'state_code',
    'state_code_y': 'state_abbv'
}, inplace=True)

final_race_df.head()
```

```python
final_race_df.to_csv('../../../machine_learning/Resources/race_data.csv', index=False)
```
